```python
# check the missing values
dataset.isnull().sum()
```

```python
dataset['CRIM'].fillna(dataset['CRIM'].median(), inplace=True)
dataset['ZN'].fillna(dataset['ZN'].mean(), inplace=True)
dataset['INDUS'].fillna(dataset['INDUS'].median(), inplace=True)
dataset['CHAS'].fillna(dataset['CHAS'].mode()[0], inplace=True)
dataset['AGE'].fillna(dataset['AGE'].median(), inplace=True)
dataset['LSTAT'].fillna(dataset['LSTAT'].median(), inplace=True)
```

```python
#  INDEPENDENT AND DEPENDENT FEATURES
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
X.head()
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
# standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
# print coefficinets and intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# on which parameters the model has been trained
regression.get_params()
```

```python
# prediction with test data
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
# plot a scatter plot for the prediction
plt.scatter(y_test, reg_pred)
```

```python
# residuls
residuals = y_test - reg_pred
```

```python
residuals
```

```python
# plot this residuals
sns.displot(residuals, kind='kde')
```

```python
# scatter plot with respect to predictions and residuals
# uniform distribution
plt.scatter(reg_pred, residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
# R square and adjusted R square
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```
